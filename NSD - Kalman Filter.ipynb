```python
alpha = .05
def Kernel(t1, t2):
    global alpha
    
    exp1 = np.exp(-alpha * abs(t1 - t2))
    exp2 = np.exp(-alpha * (t1 + t2))
    return alpha * min(t1, t2) - .5 * (exp1 - exp2)
```

```python
from scipy.interpolate import interp1d, LinearNDInterpolator
import math
dates = [pd.to_datetime(x) for x in coupons.date.unique()]
dates.sort() # все даты будущих выплат купонов/номиналов
payments = pd.DataFrame( np.zeros((len(dates), len(eurobonds_info))), columns = eurobonds_prices.columns, index = coupons['date'].sort_values().unique())
for index, row in eurobonds_info.iterrows():

    payments[row['isin']][row['maturity_date']] = payments[row['isin']][row['maturity_date']] + 100
for date in list(payments.index):
    for index, row in coupons[coupons.date == date].iterrows():
        payments[str(row['isin'])][date] += row['rate'] / 2
l = len(dates)
n = len(list(eurobonds_prices.columns))

_lambda = 10000
omega = .054
# What will be in operation:
# t - vector of points of time in future, for which discount factor are estimated, 1 x l
# discounts - discount factors for point of time from vector t, l x 1
# P - covariance matrix of system (discount function) error, l x l
first_date = list(eurobonds_prices.index)[0]
durations = list(eurobonds_durations.loc[first_date].sort_values().values)
ytms = np.array(list(eurobonds_ytms.loc[first_date].sort_values().values))
discount_factors = [np.exp(-(ytm/100)*ttm) for ytm, ttm in zip(ytms, durations)]
discount_factors = [1] + discount_factors
durations = [0] + durations
t = [(pd.to_datetime(x) - list(eurobonds_prices.index)[0]).days/365 for x in dates]
disc_func = interp1d(durations, discount_factors, kind='cubic')
initial_discounts = [ disc_func(ttm) if ttm <= max(durations) else disc_func(max(durations)) for ttm in t]


# prices = prices.fillna(0)
# nkds = nkds.fillna(0)



# Initialization of discount factors and its covariation matrix (P)

t = [(pd.to_datetime(x) - list(eurobonds_prices.index)[0]).days/365 for x in dates]
discounts = np.array([ disc_func(ttm) if ttm <= max(durations) else disc_func(max(durations)) for ttm in t])
P = np.zeros((l, l))
discounts_omega = np.array([np.exp(-omega*ttm) for ttm in t])
discounts = np.ones(len(t))
# initial_discounts = [ disc_func(ttm) if ttm <= max(durations) else disc_func(max(durations)) for ttm in t]
# payments_it = discounts_omega @ payments_it
# print(payments_it)

for date in list(eurobonds_prices.index):
    print(date)
    
    today_isins_list = eurobonds_durations.loc[date][eurobonds_durations.loc[date] > .5].index
    payments_it = payments[today_isins_list]
    prices = eurobonds_prices[today_isins_list]
    nkds = eurobonds_nkds[today_isins_list]
    
    
    dates_new = []
    for cur_date in dates:
        if cur_date > pd.to_datetime(date): dates_new = dates_new + [cur_date]
    dates = dates_new
    payments_it = payments_it.loc[payments_it.index > date]
    t = [(pd.to_datetime(x) - date).days/365 for x in dates]
    t_interpolation = np.array([0] + t)
    t = np.array(t)
    if len(t) != len(discounts): 
        discounts = discounts[1:]
        P = P[1:,1:]
        l -= 1
    discounts_interpolation = np.array([1] + list(discounts))
#     print(len(discounts), len(t), len(discounts_interpolation), len(t_interpolation))
    print(t[0])

    discount_function = interp1d(np.transpose(t_interpolation), discounts_interpolation)
    discount_interpolated = discount_function(1/365)
    discounts_omega = np.diag([np.exp(-omega*ttm) for ttm in t])
    payments_it_omega = discounts_omega @ payments_it
#     print(discount_interpolated[:3])
    Q = [[((1 / 365) / (alpha ** 2)) * Kernel(x, y) for y in t] for x in t]     
    #discounts and P prediction:
    discounts = 1 + discounts - discount_interpolated
    P = P + Q
    
    # Filter calculation
#     discount_multiples = np.array([1/discount for discount in discounts])
    error = prices.loc[date] + nkds.loc[date] - np.dot(discounts, payments_it_omega)
    today_isins = len(isins_list) - prices.loc[date].isna().sum()
    errorsKF['mae'][datetime.strftime(date, '%Y-%m-%d')] = sum(abs(error)) / today_isins
    errorsKF['rmse'][datetime.strftime(date, '%Y-%m-%d')] = (sum(error ** 2) / today_isins) ** .5
    print(np.dot(discounts, payments_it_omega))
    print(error)


    print(sum(error ** 2))
    
    R = _lambda * np.diag(eurobonds_spreads[today_isins_list].loc[date] / 2)
    S = np.dot(np.transpose(payments_it_omega), P) @ payments_it_omega + R 
    K = (P @ payments_it_omega) @ np.linalg.inv(S)
#     print(np.linalg.inv(S))

    
    #Estimate correction
#     print(np.dot(error, np.transpose(K)))
    discounts = discounts + np.dot(error, np.transpose(K))
    P = (np.eye(l) - np.dot(K, payments_it_omega.T)) @ P
#     P = P - np.dot(K, payments_it_omega.T) @ P
    
#     print(discounts)
    zcb_rates = [(discount_factor ** ttm) ** (-1) - 1 for discount_factor, ttm in zip(discounts, t)]

    discounts_dict[datetime.strftime(date, '%Y-%m-%d')] = discounts
    curves_dict[datetime.strftime(date, '%Y-%m-%d')] = np.asarray([-(np.exp(-omega) * (discount_factor ** (ttm ** -1)) - 1) * 100 for discount_factor, ttm in zip(discounts, t)])
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(t, discounts, label='d(t)')
axes.plot(t, [np.exp(omega*ttm) for ttm in t], label = 'Omega')

# axes.plot(t, initial_discounts)
# zcb_rates = [(-math.log(discount_factor) / ttm) * 100 for discount_factor, ttm in zip(discounts, t)]
zcb_rates_nsd = [(np.exp(-omega) * (discount_factor ** (ttm ** -1)) - 1) * 100 for discount_factor, ttm in zip(discounts, t)]

zcb_rates = [(-math.log(discount_factor * np.exp(omega * ttm)) / ttm) * 100 for discount_factor, ttm in zip(discounts, t)]



axes.grid()
axes.set_ylim([.5, 1.8])

t = [(pd.to_datetime(x) - date).days/365 for x in dates]
t_interpolation = np.array([0] + t)
t = np.array(t)
discounts_interpolation = np.array([1] + list(discounts))
discount_function = interp1d(np.transpose(t_interpolation), discounts_interpolation)
discount_interpolated = discount_function(1/365)
print(error)

Discounts = [np.exp(-omega*ttm) * discount_factor for discount_factor, ttm in zip(discounts, t)]
axes.plot(t, Discounts, label='D(t)')
axes.legend()

```
